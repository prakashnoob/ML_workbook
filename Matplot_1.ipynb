```python
import matplotlib.pyplot as mat
```

```python
mat.plot(2,4,6,4)
mat.ylabel("Num")
mat.xlabel("indices")
mat.title("plot")
mat.show()
```

```python
mat.plot([2,4,6,4])
mat.ylabel("Num")
mat.xlabel("indices")
mat.title("plot")
mat.show()
```

```python
mat.plot([2,4,6,4])
mat.ylabel("Num")
mat.xlabel("indices")
mat.title("plot")
mat.grid()
mat.show()
```

```python
mat.plot([1,2,3,4],[9,8,7,6],'ro')
mat.grid()
mat.show()
```

```python
import numpy as np
p=np.arange(0.,5.,0.2)
mat.plot(p,p**2, 'b--',label='^2')
mat.plot(p,p**2.2,'rs',label='^2.2')
mat.plot(p,p**2.5,'g^',label='^2.5')
mat.grid()
mat.legend()
mat.show()
```
